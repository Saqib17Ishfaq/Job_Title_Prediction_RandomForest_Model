```python
import pandas as pd

def predict_job_from_user_input(model_pipeline):
    print("\n=== Job Prediction System ===")
    print("Please enter the following details:\n")
    
    # Collect user input for each feature
    input_data = {}
    
    # Personal Information
    input_data['Gender'] = [input("Gender (Male/Female/Other): ").strip()]
    
    # Education Details
    input_data['Course in UG'] = [input("Undergraduate Course (e.g., B.E, B.Tech, B.Sc): ").strip()]
    input_data[' UG specialization? Major Subject (Eg; Mathematics)'] = [
        input(" UG Specialization (e.g., Computer Science Engineering): ").strip()
    ]
    
    # Certification
    input_data['Certificate course title'] = [
        input("Certificate/Degree (e.g., Microsoft certification): ").strip()
    ]
    
    # Academic Performance
    while True:
        try:
            cgpa = float(input("Average CGPA/Percentage in UG (e.g., 66.5): "))
            input_data['Average CGPA or Percentage obtained in under graduation'] = [cgpa]
            break
        except ValueError:
            print("Please enter a valid number (e.g., 66.5 or 8.5)")
    
    # Skills and Interests
    input_data['Interests'] = [input("Your professional interests (e.g., Technology): ").strip()]
    input_data['Skills'] = [
        input("Your skills (comma separated, e.g., Critical Thinking, SQL): ").strip()
    ]
    
    # Work Experience
    input_data['Work in the past'] = [
        input("Previous work experience (e.g., IT Specialist): ").strip()
    ]
    
    # Convert to DataFrame
    input_df = pd.DataFrame(input_data)
    
    # Make prediction
    try:
        prediction = model_pipeline.predict(input_df)[0]
        probabilities = model_pipeline.predict_proba(input_df)[0]
        
        print("\n=== Prediction Result ===")
        print(f"Predicted Job Title: {prediction}")
        
        print("\nPrediction Probabilities:")
        for job, prob in zip(model_pipeline.classes_, probabilities):
            print(f"{job}: {prob:.2%}")
        
        return prediction
        
    except Exception as e:
        print(f"\nError making prediction: {str(e)}")
        return None

# Example usage (assuming you have a trained model_pipeline)
if __name__ == "__main__":
    # First you need to have your trained model_pipeline
    # model_pipeline = your_trained_model
    
    print("Job Prediction System")
    while True:
        predicted_job = predict_job_from_user_input(model_pipeline)
        
        another = input("\nMake another prediction? (y/n): ").lower()
        if another != 'y':
            print("\nThank you for using the Job Prediction System!")
            break
```
